```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Datasets preprocessing

```python
df = pd.read_csv(r"C:\Users\ukasy\project ML\Study Group\Prjocet ML tebak kelulusan mahasiswa\Kelulusan Train.csv")

df.sample(10)
```

```python
df = df.drop(['JENIS KELAMIN', 'STATUS MAHASISWA', 'STATUS NIKAH', 'NAMA'], axis=1,)

df = df.dropna(axis=0)
df.duplicated().sum()

df.sample(3)
print(df['STATUS KELULUSAN'])
print(df['IPK '])
```

1. Histogram
histogram ini fungsinya adalah untuk melihat persebaran masing2 kolom baik feature dan target pada data set

Visualisasi Data

```python
for column in df.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Histogram {column}')
    plt.xlabel(f'{column}')
    plt.ylabel('Frekuensi')
    plt.show()
```

2. Pie Chart
Pie chart ini berfungsi untuk melihat persentasi persebaran data pada datasets (hanya saja untuk data ini kurg cocok karena terlalu banyak perbedaan trutama pada IPS dan IPK)

```python
for column in df.columns:
    counts = df[column].value_counts()
    plt.figure(figsize=(8, 6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Pie Chart {column}')
    plt.axis('equal')  # Agar pie chart menjadi lingkaran
    plt.show()
```

3. Scater plot Kelulusan vs IPK
Scater Plot digunakan untuk membandingkan persebaran dengan 2 variabel datasets (sebenarnya bsa lebih menggunakan PCA, tpi saya blom paham wkwk) pada koordinat kartesius. Scatter ini sangan sering digunakan trutama pada regresi

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['IPK '], df['STATUS KELULUSAN'])
plt.title('IPK vs Status Kelulusan')
plt.xlabel('IPK')
plt.ylabel('Status Kelulusan')
plt.grid(True)
plt.show()
```
